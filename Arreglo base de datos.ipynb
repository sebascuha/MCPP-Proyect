```python
#Función para unificar nombre de departamentos
def arreglo_nombres(lista_departamentos):
    for row in range(len(lista_departamentos)):
        if lista_departamentos[row] == ('VAUPES'):
            lista_departamentos[row] = 'VAUPÉS'
        elif lista_departamentos[row] == ('ARCHIP.DE SAN ANDRES, PROVIDENCIA Y SANTA CATALINA'):
            lista_departamentos[row] = 'SAN ANDRÉS, PROVIDENCIA Y SANTA CATALINA (ARCHIPIÉLAGO)'
        elif lista_departamentos[row] == ('ARCHIPIÉLAGO DE SAN ANDRÉS, PROVIDENCIA Y SANTA CATALINA'):
            lista_departamentos[row] = 'SAN ANDRÉS, PROVIDENCIA Y SANTA CATALINA (ARCHIPIÉLAGO)'
        elif lista_departamentos[row] == 'NARIÃ±O':
            lista_departamentos[row] = 'NARIÑO'        
        elif lista_departamentos[row] == 'QUINDIO':
            lista_departamentos[row] = 'QUINDÍO'
        elif lista_departamentos[row] == 'GUAINIA':
            lista_departamentos[row] = 'GUAINÍA'
        elif lista_departamentos[row] == 'ATLÁNTICO':
            lista_departamentos[row] = 'VAUPÉS'
        elif lista_departamentos[row] == ('BOGOTÁ, D.C.'):
            lista_departamentos[row] = 'BOGOTÁ D.C.'
        elif lista_departamentos[row] == ('BOGOTA, D.C.'):
            lista_departamentos[row] = 'BOGOTÁ D.C.'
        elif lista_departamentos[row] == 'BOLIVAR':
            lista_departamentos[row] = 'BOLÍVAR'
        elif lista_departamentos[row] == 'BOYACA':
            lista_departamentos[row] = 'BOYACÁ'
        elif lista_departamentos[row] == 'CHOCO':
            lista_departamentos[row] = 'CHOCÓ'
        elif lista_departamentos[row] == 'CORDOBA':
            lista_departamentos[row] = 'CÓRDOBA'
        elif lista_departamentos[row] == 'CAQUETA':
            lista_departamentos[row] = 'CAQUETÁ'
        elif lista_departamentos[row] == 'ATLANTICO':
            lista_departamentos[row] = 'ATLÁNTICO'
        else: 
            next 
    return lista_departamentos 
```

```python
# anted de aplicar la funcion para arreglo nombres de departamento
subdf_merged2['depto'].unique()
```

```python
# Aplicando función
subdf_merged2['depto'] = arreglo_nombres(subdf_merged2['depto'])
```

```python
# Despues de haber aplicado cambios en los nombres de los departamentos
subdf_merged2['depto'].unique()
```

```python
Vic_desplazamiento_forzado_edit['depto'] = arreglo_nombres(Vic_desplazamiento_forzado_edit['depto'])
PIB_depto_pc_edit['depto'] = arreglo_nombres(PIB_depto_pc_edit['depto'])
```
